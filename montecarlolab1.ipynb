```python
df = pd.DataFrame(columns = ['X0','a,b,m','Numbers','Distinct values before repitition'] )
```

```python

for x0 in range(0,11):
    numbers = []
    x = (6*x0)%11
    
    t = x
    numbers.append(t)
    x = (6*x)%11
    while x!=t:
        numbers.append(x)
        x = (6*x)%11
    distinct = len(numbers)
    row = {'X0':x0,'a,b,m':'6,0,11','Numbers':tuple(numbers),'Distinct values before repitition':distinct}
    df = df.append(row,ignore_index=True)
    
for x0 in range(0,11):
    numbers = []
    x = (3*x0)%11
    
    t = x
    numbers.append(t)
    x = (3*x)%11
    while x!=t:
        numbers.append(x)
        x = (3*x)%11
    distinct = len(numbers)
    row = {'X0':x0,'a,b,m':'3,0,11','Numbers':tuple(numbers),'Distinct values before repitition':distinct}
    df = df.append(row,ignore_index=True)
df   
```

## Ques 2

Dataframe having frequency of values in all ranges for different values of x

```python
df2 = pd.DataFrame(columns= ['x0','Range','Frequency'])
values_of_x0 = [1,2,3,4,5]

def count(list1, l, r):
    c = 0
    # traverse in the list1
    for x in list1:
        # condition check
        if x>= l and x< r:
            c+= 1
    return c

list_of_list = []
for xx in values_of_x0:
    ui = []
    x0 = xx
    x = (1597*x0+51749)%244944
    u = x/244944
    ui.append(u)
    for i in range (1,10001):
        x = (1229*x+1)%2048
        u = x/2048
        ui.append(u)

    f = []
    g = []
    i =0.00
    while i<1:
        k = count(ui,i,i+0.05)
        f.append(k)
        e = str(round(i,2))+'-'+str(round(i+0.05,2))
        g.append(e)
        i = i+0.05
        row2 = {'x0':xx,'Range':e,'Frequency':k}
        df2 = df2.append(row2,ignore_index=True)
    list_of_list.append(f)

list_of_list
```

Scatter plots for the same(I found that scatter plots give better intuition of what is happening for different values of x0, so i chose this)


```python
figure, ax = plt.subplots()
ax.scatter(g,list_of_list[0],c ='red',label = 'x0 =1',s=100)
ax.scatter(g,list_of_list[1],c ='blue',label = 'x0 =2',s=80)
ax.scatter(g,list_of_list[2],c ='green',label = 'x0 =3',s=60)
ax.scatter(g,list_of_list[3],c ='yellow',label = 'x0 =4',s=40)
ax.scatter(g,list_of_list[4],c ='black',label = 'x0 =5',s=20)

plt.xticks(rotation = 'vertical')
ax.legend()
plt.ylabel('Frequency')
plt.xlabel('Ranges')
plt.show()
```
